#W4A6
"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import math
def prime(n):
    if n>1:
        for i in range(2,int(math.sqrt(n)+1)):
            if n%i==0:
                return False
        return True

tong=0
a,b=map(int,input().split())
for j in range(a,b+1):
    if prime(j):
        tong+=j
print(tong)
```

#W4A7
"[LargestPrimeFactor]
Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n." 

```python
import math
def prime(n):
    if n>1:
        for i in range(2,int(math.sqrt(n)+1)):
            if n%i==0:
                return False
        return True

n=int(input())
max=1
for i in range(1,n+1):
    if prime(i) and n%i==0:
        max=i
print(max)

```

[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).

```python
import math

n = int(input())
tim_thay=False

for i in range(0, int(math.sqrt(n)) + 1):
    binh_phuong = i**2
    chuoi_so = str(binh_phuong)
    
    if len(set(chuoi_so)) == len(chuoi_so):
        print(binh_phuong, end=" ")
        tim_thay=True
if not tim_thay:
    print("(no number)")
```

W4A13 Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a,b=map(int,input().split())
tong1=0
tong2=0
for i in range(1,a):
    if a%i==0:
        tong1+=i
for j in range(1,b):
    if b%j==0:
        tong2+=j
if tong1==b and tong2==a:
    print(True)
else:
    print(False)
```
