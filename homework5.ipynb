Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x. Một số được gọi là số Armstrong nếu tổng các lũy thừa bậc n của từng chữ số của nó bằng chính số đó, trong đó n là số chữ số của số đó.


```python
def armstrong(n):
    tong=0
    dem=0
    x=n
    z=n
    while n>0:
        dem+=1
        n=n//10

    while x>0:
        k=x%10
        tong=tong+k**dem
        x=x//10

    if tong==z:
        return True
    else:
        return False
a=int(input())
if armstrong(a):
    print(True)
else:
    print(False)
    
        
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n<=0 trả về 0

```python
def so_khac_nhau(n):
    while n>0:
        k1=n%10
        n=n//10
        k2=n%10
        if k2==k1:
            return False
        else:
            continue
    return True

def sum_1_to_n(n):
    tong=0
    if n<=0:
        return 0
    for x in range(1,n+1):
        if so_khac_nhau(x):
            tong+=x
    return tong
print(sum_1_to_n(0))
```
