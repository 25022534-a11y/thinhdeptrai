BÀI 1: Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
import math
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    tien=0
    if 0<km<=0.8:
        if car_type==4:
            tien+=11000
        elif car_type==7:
            tien+=12000
    else:
        if car_type==4:
            if 0.8<km<=20:
                tien+=13000*(km-0.8)+11000
            elif km>20:
                tien+=260600+10500*(km-20)

        elif car_type==7:
            if 0.8<km<=20:
                tien+=14000*(km-0.8)+12000
            elif km>20:
                tien+=280800+12000*(km-20)
    if wait_minutes>5:
        tien+=(math.ceil(wait_minutes-5)//5)*2000
    
    if start_hour in [22,23,0,1,2,3,4,5]:
        tien+=tien*0.2

    if km>30 and start_hour not in [22,23,0,1,2,3,4,5]:
        tien=tien-tien*0.1
    
    if tien<50000 and km>5:
        tien=50000

    return int(tien),"VNĐ"

car_type=int(input())
start_hour=int(input())
km=float(input())
wait_minutes=int(input())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

BÀI 2: Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False.
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng. 


```python
def check(x,n):
    tong=0
    for i in range(1,n+1):
        tong+=i
    if tong==x:
        return True
    return False
print(check(10,6))
```

sinh sớm hơn), 2 nếu người 2 lớn tuổi hơn, và 0 nếu bằng nhau.
Ví dụ: 27 2 2007 so với 28 2 2007 → trả về 1 vì ngày 27 sớm hơn nên người 1 lớn tuổi hơn. 


```python
def check(day1, month1, year1, day2, month2, year2):
    if year1<year2:
        return 1
    elif year1>year2:
        return 2
    else:
        if month1<month2:
            return 1
        elif month1>month2:
            return 2
        else:
            if day1<day2:
                return 1
            elif day1>day2:
                return 2
    
print(check(31,2,2006,30,2,2007))
```

BÀI 9 Yêu cầu: Cho số nguyên n, trả về:
- 5n + 1 nếu n mod 3 == 1
- 6n + 2 nếu n mod 3 == 2
- 7n + 3 nếu n mod 3 == 0 


```python
def tinh_gia_tri(n):
    if n % 3 == 1:
        return 5*n + 1
    elif n % 3 == 2:
        return 6*n + 2
    else: 
        return 7*n + 3
```
